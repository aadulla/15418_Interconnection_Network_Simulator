```python
test_msg_0 = [ 
             path1+'test_16', # SF + packet
             path1+'test_20', # SF + flit 
             path1+'test_24', # CT + packet
             path1+'test_28', # CT + flit
             ]
test_msg_1 = [ 
             path1+'test_17', # SF + packet
             path1+'test_21', # SF + flit 
             path1+'test_25', # CT + packet
             path1+'test_29', # CT + flit
             ]
test_msg_2 = [ 
             path1+'test_18', # SF + packet
             path1+'test_22', # SF + flit 
             path1+'test_26', # CT + packet
             path1+'test_30', # CT + flit
             ]
test_msg_3 = [ 
             path1+'test_19', # SF + packet
             path1+'test_23', # SF + flit 
             path1+'test_27', # CT + packet
             path1+'test_31', # CT + flit
             ]

global_stats_df_dict_0 = data_parser(test_msg_0)
global_stats_df_dict_1 = data_parser(test_msg_1)
global_stats_df_dict_2 = data_parser(test_msg_2)
global_stats_df_dict_3 = data_parser(test_msg_3)

global_stats_df_dict_lst = [global_stats_df_dict_0, global_stats_df_dict_1, global_stats_df_dict_2, global_stats_df_dict_3]

titles = [
            'Routing: Mesh Adaptive, Message Size: 10, Number of Messages: 5000',
            'Routing: Mesh Adaptive, Message Size: 50, Number of Messages: 1000',
            'Routing: Mesh Adaptive, Message Size: 100, Number of Messages: 500',
            'Routing: Mesh Adaptive, Message Size: 500, Number of Messages: 100',
         ]

legend = ["SF + Packet", "SF + Flit", "CT + Packet", "CT + Flit"]
```

```python
tmp_dict = {"SF + Flit": [], "CT + Flit": [], "SF + Packet": [], "CT + Packet": []}
for stats_dict in global_stats_df_dict_lst:
    for val, header in zip(stats_dict['aggregate'].loc['throughput'], legend):
        tmp_dict[header].append(val)
df = pd.DataFrame(tmp_dict, index=["Traffic Pattern 1", 
                                   "Traffic Pattern 2", 
                                   "Traffic Pattern 3", 
                                   "Traffic Pattern 4"])
ax = df.plot.bar(rot=0, logy=True)
ax.set_ylabel("Throughput (msgs/cycle)")
ax.set_title("Mesh Adaptive: Throughput vs Traffic Pattern")
fig = ax.get_figure()
fig.set_size_inches(8, 5)
```
