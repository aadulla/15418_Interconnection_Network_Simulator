```python
path1 = 'test_suite/routing_+_flow_control_+_message_size_+_message_distribution/'
```

```python
path2 = 'test_suite/routing_+_buffer_capacity_+_virtual_channel_+_granularity/'
```

```python
test_msg_0 = [ 
             path1+'test_0', # SF + packet
             path1+'test_4', # SF + flit 
             path1+'test_8', # CT + packet
             path1+'test_12', # CT + flit
             ]
test_msg_1 = [ 
             path1+'test_1', # SF + packet
             path1+'test_5', # SF + flit 
             path1+'test_9', # CT + packet
             path1+'test_13', # CT + flit
             ]
test_msg_2 = [ 
             path1+'test_2', # SF + packet
             path1+'test_6', # SF + flit 
             path1+'test_10', # CT + packet
             path1+'test_14', # CT + flit
             ]
test_msg_3 = [ 
             path1+'test_3', # SF + packet
             path1+'test_7', # SF + flit 
             path1+'test_11', # CT + packet
             path1+'test_15', # CT + flit
             ]

global_stats_df_dict_0 = data_parser(test_msg_0)
global_stats_df_dict_1 = data_parser(test_msg_1)
global_stats_df_dict_2 = data_parser(test_msg_2)
global_stats_df_dict_3 = data_parser(test_msg_3)

global_stats_df_dict_lst = [global_stats_df_dict_0, global_stats_df_dict_1, global_stats_df_dict_2, global_stats_df_dict_3]

titles = [
            'Routing: Mesh XY, Message Size: 10, Number of Messages: 5000',
            'Routing: Mesh XY, Message Size: 50, Number of Messages: 1000',
            'Routing: Mesh XY, Message Size: 100, Number of Messages: 500',
            'Routing: Mesh XY, Message Size: 500, Number of Messages: 100',
         ]

legend = ["SF + Packet", "SF + Flit", "CT + Packet", "CT + Flit"]
```

```python
fig, ax = plt.subplots(2,2)
time_series_subplot(ax, 
                    global_stats_df_dict_lst, 
                    'tx',
                    "Clock Cycle", 
                    "Number of Flits Transmitted", 
                    titles, 
                    legend, 
                    200)
fig.set_size_inches(15, 11)
fig.suptitle("Flit Transmission Time Series Plot")
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
time_series_subplot(ax, 
                    global_stats_df_dict_lst, 
                    'rx',
                    "Clock Cycle", 
                    "Number of Flits Received", 
                    titles, 
                    legend, 
                    200)
fig.set_size_inches(15, 11)
fig.suptitle("Flit Reception Time Series Plot")
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
time_series_subplot(ax, 
                    global_stats_df_dict_lst, 
                    'buffers',
                    "Clock Cycle", 
                    "Buffer Efficiency", 
                    titles, 
                    legend, 
                    200)
fig.set_size_inches(15, 11)
fig.suptitle("Buffer Efficiency Time-Series Plot")
plt.show()
```
