```python
path2 = 'test_suite/routing_+_buffer_capacity_+_virtual_channel_+_granularity/'
```

```python
test_msg_0 = [ 
             path1+'test_0', # SF + M_XY
             path1+'test_16', # SF + M_Adapt
             path1+'test_8', # CT + M_XY
             path1+'test_24', # CT + M_Adapt
             ]
test_msg_1 = [ 
             path1+'test_1', # SF + M_XY
             path1+'test_17', # SF + M_Adapt
             path1+'test_9', # CT + M_XY
             path1+'test_25', # CT + M_Adapt
             ]
test_msg_2 = [ 
             path1+'test_2', # SF + M_XY
             path1+'test_18', # SF + M_Adapt
             path1+'test_10', # CT + M_XY
             path1+'test_26', # CT + M_Adapt
             ]
test_msg_3 = [ 
             path1+'test_3', # SF + M_XY
             path1+'test_19', # SF + M_Adapt
             path1+'test_11', # CT + M_XY
             path1+'test_27', # CT + M_Adapt
             ]

global_stats_df_dict_0 = data_parser(test_msg_0)
global_stats_df_dict_1 = data_parser(test_msg_1)
global_stats_df_dict_2 = data_parser(test_msg_2)
global_stats_df_dict_3 = data_parser(test_msg_3)

global_stats_df_dict_lst = [global_stats_df_dict_0, global_stats_df_dict_1, global_stats_df_dict_2, global_stats_df_dict_3]

titles = [
            'Granularity: Packet, Message Size: 10, Number of Messages: 5000',
            'Granularity: Packet, Message Size: 50, Number of Messages: 1000',
            'Granularity: Packet, Message Size: 100, Number of Messages: 500',
            'Granularity: Packet, Message Size: 500, Number of Messages: 100',
         ]

legend = ["M_XY + SF", "M_Adapt + SF", "Mesh_XY + CT", "M_Adapt + CT"]
```

```python
fig, ax = plt.subplots(2,2)
time_series_subplot(ax, 
                    global_stats_df_dict_lst, 
                    'tx',
                    "Clock Cycle", 
                    "Number of Flits Transmitted", 
                    titles, 
                    legend, 
                    200)
fig.set_size_inches(15, 11)
fig.suptitle("Flit Transmission Time Series Plot")
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
time_series_subplot(ax, 
                    global_stats_df_dict_lst, 
                    'rx',
                    "Clock Cycle", 
                    "Number of Flits Received", 
                    titles, 
                    legend, 
                    200)
fig.set_size_inches(15, 11)
fig.suptitle("Flit Reception Time Series Plot")
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
time_series_subplot(ax, 
                    global_stats_df_dict_lst, 
                    'stalls',
                    "Clock Cycle", 
                    "Number of Contentions", 
                    titles, 
                    legend, 
                    100)
fig.set_size_inches(15, 11)
fig.suptitle("Contentions Time-Series Plot")
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
time_series_subplot(ax, 
                    global_stats_df_dict_lst, 
                    'buffers',
                    "Clock Cycle", 
                    "Buffer Efficiency", 
                    titles, 
                    legend, 
                    100)
fig.set_size_inches(15, 11)
fig.suptitle("Buffer Efficiency Time-Series Plot", fontsize=28)
plt.show()
```
